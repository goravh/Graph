```python
# importing basic libraries
import numpy as np
import pandas as pd

# Importing matplotlib for data visualization
import matplotlib.pyplot as plt
```

```python
# lets read the dataset

# using employee dataset
data = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')

# lets take only 200 rows as we are only going to visualize it
data = data.loc[:199,:]
print(data.shape)
```

```python
# lets check the structure of the dataset using the Head function

data.head()
```

```python
# let's check the columns present in the dataset

data.columns
```

```python
# lets check the documentation of the line plots available in matplotlib
help(plt.plot)
```

```python
# basic line plot

plt.plot(data['Department'],
         data['Age'])
```

```python
# lets try to aggregate the age using mean in each dept.

x = data.pivot_table(index = 'Department',
                     values = 'Age',
                     aggfunc = 'mean').plot(kind='line')
```

```python
# lets try to aggregate the age using mean in each dept.

# to know more about pivot tables, refer to Pandas->Pivot tables tutorial

x = data.pivot_table(index = 'Department',
                     values = ['Age','TotalWorkingYears'],
                     aggfunc = 'mean').plot(kind='line')
```

```python
# trying to plot the line plots in a different plots

data.pivot_table(index = 'Department',
                     values = ['Age'],
                     aggfunc = 'mean').plot(kind='line', color='red')

data.pivot_table(index = 'Department',
                     values = ['TotalWorkingYears'],
                     aggfunc = 'mean').plot(kind='line', color='blue')


data.pivot_table(index = 'Department',
                     values = ['HourlyRate'],
                     aggfunc = 'mean').plot(kind='line', color='green')



data.pivot_table(index = 'Department',
                     values = ['DailyRate'],
                     aggfunc = 'mean').plot(kind='line', color='yellow')

plt.show()
```

```python
# Trying to plot line plots in different plots Using Subplots

plt.rcParams['figure.figsize'] = (12, 5)

plt.subplot(221)
x = data.pivot_table(index = 'Department',
                     values = ['Age'],
                     aggfunc = 'mean')
plt.plot(x,color='red')
plt.ylabel('Age')

plt.subplot(222)
y = data.pivot_table(index = 'Department',
                     values = ['TotalWorkingYears'],
                     aggfunc = 'mean')
plt.plot(y, color = 'blue')
plt.ylabel('Experience')

plt.subplot(223)
z = data.pivot_table(index = 'Department',
                     values = ['HourlyRate'],
                     aggfunc = 'mean')
plt.plot(z, color ='green')
plt.ylabel('Salary')

plt.subplot(224)
a = data.pivot_table(index = 'Department',
                     values = ['PerformanceRating'],
                     aggfunc = 'mean')
plt.plot(a, color = 'yellow')
plt.ylabel('Performance')

plt.suptitle('Line Plots using Subplots')
plt.show()
```

```python
# lets use different patterns for the lines

plt.rcParams['figure.figsize'] = (12, 5)

plt.subplot(221)
x = data.pivot_table(index = 'Department',
                     values = ['Age'],
                     aggfunc = 'mean')
plt.plot(x, color='red')
plt.ylabel('Age')

plt.subplot(222)
y = data.pivot_table(index = 'Department',
                     values = ['TotalWorkingYears'],
                     aggfunc = 'mean')
plt.plot(y, '-+', color = 'blue')
plt.ylabel('Experience')

plt.subplot(223)
z = data.pivot_table(index = 'Department',
                     values = ['HourlyRate'],
                     aggfunc = 'mean')
plt.plot(z, '--',  color ='green')
plt.ylabel('Salary')

plt.subplot(224)
a = data.pivot_table(index = 'Department',
                     values = ['PerformanceRating'],
                     aggfunc = 'mean')
plt.plot(a, '-*', color = 'black')
plt.ylabel('Performance')

plt.suptitle('Using Different Patterns for Different Line Plots')
plt.show()
```
