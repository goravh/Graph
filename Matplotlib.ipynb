```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
# Scatter plot
```

```python
x = np.arange(0,20)
y = np.arange(30,50)
```

```python
plt.scatter(x,y, c= "g", marker = "*")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title('Scatter plot using Matplotlib')
            
plt.show()            
```

```python
# line plot

plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title('Line Graph using Matplotlib')
            
plt.show()            
```

```python
df = [20,15,25,30,10,100]
```

```python
plt.boxplot(df)
plt.show()
```

```python
# class
x = ['8th','9th','10th']
# strength
y = [45,50,49]
```

```python
plt.bar(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title('Line Graph using Matplotlib')
plt.show()            
```

```python
# Pie Chart
```

```python
x = [1000, 200, 300, 50]
labels = ['China', 'US', 'Russia', 'India']
colors = ['r', 'g', 'b', 'y']
```

```python
# without explode
plt.pie(x, labels = labels, colors=colors, shadow = True)
plt.title("This is as PIE Chart")
plt.show()
```

```python
# with explode
e = (0.1,0,0,0)
plt.pie(x,explode = e, labels = labels, colors=colors, autopct = "%1.3f%%" , shadow = True)
plt.title("This is as PIE Chart")
plt.show()
```
