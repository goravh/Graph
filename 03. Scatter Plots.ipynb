```python
# importing the basic libraries 
import numpy as np
import pandas as pd

# importing matplotlib for data visualization
import matplotlib.pyplot as plt
```

```python
# lets import the dataset
data = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')

# lets check the shape of the dataset
data.shape
```

```python
# lets also check the structure of the dataset
data.head()
```

```python
# lets check the documentation of scatter plot
help(plt.scatter)
```

```python
# lets explore a simple scatter plot

# setting the size of the plot
plt.rcParams['figure.figsize'] = (9, 5)

# setting the parameters of the plot
plt.scatter(data['Age'],
            data['MonthlyRate'])
plt.show()
```

```python
# second example for exploring scatter plot with adding labels and titles

# setting the parameters in the plota
# also, specifying a different color for the plot
plt.scatter( data['TotalWorkingYears'],
            data['MonthlyRate'],
            color = 'red')

# specifying the xlabel, ylabel, and the title of the plot
plt.ylabel('Monthly Rate')
plt.xlabel('Total Working Years')
plt.title('Comparing Monthly Wages with Experience')
plt.show()
```

```python
# second example for exploring scatter plot with adding labels and titles

# setting the size of the plot
plt.rcParams['figure.figsize'] = (16, 6)

# setting up the parameters of the plot
# specifying the third size variable in s
plt.scatter( data['TotalWorkingYears'],
            data['MonthlyRate'],
            s = data['DistanceFromHome'],
            color = 'green')

plt.ylabel('Monthly Rate')
plt.xlabel('Total Working Years')

# specifying the fontsize of the title of the plot
plt.title('Comparing Monthly Wages with Experience', fontsize = 20)
plt.show()
```
